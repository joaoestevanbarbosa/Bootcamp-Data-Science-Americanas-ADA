1) Utilizando o DataFrame importado (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)

```python
import pandas as pd
import numpy as np

ler_aluno3 = pd.read_csv(r"C:\Users\joao_\Desktop\Modal DS\tecprog1\dados\alunos3.csv", sep = ';', decimal = ',')
ler_aluno3
```

```python
#iloc[linha 0 ate final, coluna 4 até final]
# .sum(), axis = 0 compara por linhas, axis = 1 compara por coluna

ler_aluno3['Soma'] = ler_aluno3.iloc[:, -4:].sum(axis=1)

ler_aluno3
```

```python
ler_aluno3['Media'] = (ler_aluno3['Soma'])/ler_aluno3.iloc[:, -4:].count(axis=1)

ler_aluno3
```

2) Quem foram os alunos que obtiveram a maior e a menor média

```python
filtro_maior_media = ler_aluno3['Media'] == ler_aluno3['Media'].max()
ler_aluno3[filtro_maior_media]
```

```python
filtro_menor_media = ler_aluno3['Media'] == ler_aluno3['Media'].min()
ler_aluno3[filtro_menor_media]
```

3) Agora una este dataframe com o `cadastro_alunos.xlsx`

```python
ler_cadastro_alunos = pd.read_excel(r"C:\Users\joao_\Desktop\Modal DS\tecprog1\dados\cadastro_alunos.xlsx")

ler_cadastro_alunos
```

```python
alunos_cadastrados = ler_aluno3.merge(ler_cadastro_alunos, on  = 'RA', how = 'left')

alunos_cadastrados
```

4) Qual a média entre as Media_provas dentro do público feminino? e masculino?

```python
media_sexo = alunos_cadastrados.groupby('Sexo').mean('Media')
media_sexo
```

5) Qual a média de idade das pessoas que obtiveram Media_provas maior ou igual a 7?

```python
# maior_que_sete = alunos_cadastrados['Media'] > 7
# alunos_cadastrados[maior_que_sete]

media_idade = alunos_cadastrados.query('Media >=7')['Idade'].mean()
media_idade

# media2 = alunos_cadastrados[alunos_cadastrados['Media'] >=7]['Idade'].mean()
# media2
```

6) Qual das cidades possui o maior média de Media_provas? E qual é este valor?

```python
soma_media = alunos_cadastrados[['Media','Cidade']].groupby(['Cidade']).sum()
contagem_cidade = alunos_cadastrados[['Media', 'Cidade']].groupby(['Cidade']).count()
media_cidade = soma_media / contagem_cidade

# media_cidade

media_max_cidade = media_cidade['Media'].max()
alunos_cadastrados.query('Cidade == @media_max_cidade')
```
