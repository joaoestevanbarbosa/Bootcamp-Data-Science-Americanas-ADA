```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

### Questão 2

Os exercícios de 1 a 5 serão desenvolvidos utilizando o _dataset_ `Titanic.csv`:
    
Calcule a frequência relativa, relativa percentual e acumulativa para os sobreviventes no _Titanic_.

__Dica.:__ Utilize a função `.value_counts()`

```python
# Print do Titulo
print("Tabela de frequência RELATIVA da coluna 'Survived':")

# Print da Tabela de Frequencia absoluta
display(df_titanic["alive"].value_counts(normalize=True))
```

```python
# Print do Titulo
print("Tabela de frequência RELATIVA PERCENTUAL da coluna 'Survived':")

# Print da Tabela de Frequencia Percentual
df_titanic["survived"].value_counts(normalize=True).apply(lambda x: str(round(x*100, 2)) + "%")
```

```python
# Print do Titulo
print("Tabela de frequência ACUMULATIVA da coluna 'Survived':")

# Print da Tabela de Frequencia Acumulativa
df_titanic["survived"].value_counts(normalize=True).cumsum()
```

### Questão 3

Os exercícios de 1 a 5 serão desenvolvidos utilizando o _dataset_ `Titanic.csv`:
    
Utilizando a coluna `Age` do _dataset Titanic_, defina as principais métricas estatísticas para essa variável tais como:
- Média;
- Desvio Padrão;
- Mínimo;
- Primeiro Quartil;
- Segundo Quartil (Mediana);
- Terceiro Quartil;
- Distância Interquartil (IQR);
- Máximo;
- Skewness;
- Moda.

```python
# Calculo das Metricas para AGE
print("Média de idades:", df_titanic["age"].mean())
print("Desvio padrão de idades:", df_titanic["age"].std())
print("\nIdade mínima:", df_titanic["age"].min())
print("\nIdade Q1:", df_titanic["age"].quantile(0.25))
print("Mediana de idades ou Q2:", df_titanic["age"].median())
print("Idade Q3:", df_titanic["age"].quantile(0.75))
print("IQR das idades:", df_titanic["age"].quantile(0.75) - df_titanic["age"].quantile(0.25))
print("\nIdade máxima:", df_titanic["age"].max())
print("\nSkewness das idades:", df_titanic["age"].skew())
print("\nIdade(s) mais comum(s):")
display(df_titanic["age"].mode())
```

```python
#versao resumida
df_titanic['age'].describe()
```

### Questão 4

Os exercícios de 1 a 5 serão desenvolvidos utilizando o _dataset_ `Titanic.csv`:
    
Crie um gráfico da dispersão das idades dos passageiros do _Titanic_ e identifique os pontos onde se encontram a média, mediana e moda das idades.

```python
# Define o tamanho da figura
plt.figure(figsize=(12, 6))

# Define o Titulo
plt.title("Distribuição de idades dos passageiros do Titanic", size=18)

# Plot do Histograma
sns.histplot(df_titanic["age"], kde=True, alpha=0.2)

# plotando média
plt.axvline(x=df_titanic["age"].mean(), color="red", label="média")

# plotando a mediana
plt.axvline(df_titanic["age"].median(), color="green", label="mediana")

# Loop para plotar as modas
for i in range(df_titanic["age"].mode().shape[0]):
    plt.axvline(df_titanic["age"].mode()[i], color="yellow", label="moda")

# Cria uma legenda
plt.legend()

# Mostra o Gráfico
plt.show()
```

### Questão 5

Os exercícios de 1 a 5 serão desenvolvidos utilizando o _dataset_ `Titanic.csv`:
    
Ainda trabalhando com os valores das idades crie três _Boxplots_:

```python
# Define o tamanho da figura
plt.figure(figsize=(8, 6))

# Cria um BoxPlot
sns.boxplot(data=df_titanic, x="age")

# Cria um titulo
plt.title("Boxplot das Idades")

#Mostra o gráfico
plt.show()
```

```python
# Define o tamanho da figura
plt.figure(figsize=(8, 6))

# Plot do Boxplot
sns.boxplot(data=df_titanic, y="age", x="survived", hue="sex")

# Cria um titulo
plt.title("Boxplot das Idades e Sexo por classe de sobreviventes")

# Mostra o gráfico
plt.show()
```
