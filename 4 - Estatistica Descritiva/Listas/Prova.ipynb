### Questão 3 - Rafael queria levar dinheiro na escola para comprar balas, mas sua mãe só deixou que ele levasse duas moedas para a escola. Sabendo que dentro do cofre do Rafael, temos as seguintes moedas:

10 centavos = 5 moedas

25 centavos = 3 moedas

50 centavos = 10 moedas

01 real = 2 moedas

Qual a probabilidade de Rafael retirar respectivamente uma moeda de 1 real e uma moeda de 25 centavos?

```python
#TOTAL MOEDAS
moedas = 5 + 3 + 10 + 2

#rafael tira moeda de 1 real, temos 2 moedas das 20 portanto
p1 = 2/moedas

#ao tirar 1 moeda, cai pra 19 moedas, portanto:
moedas_novo = moedas - 1

# agora tiramos uma moeda de 25 centavos, que são 3 em 19:
p2 = 3/moedas_novo

# a probabilidade é a multiplicação entre evento p1 e evento p2:
resultado = p1 * p2

str(round(resultado*100,1)) + ' %' #colocando em % e arredondando 1 casa
```

### Questão 4 - Em uma filial de um banco, 40% dos clientes que vão ao caixa eletrônico sacam dinheiro do caixa. Utilizando de uma distribuição binomial, qual é a probabilidade de no máximo 12 clientes sacarem dinheiro em um dia que tivemos 30 clientes no caixa eletrônico?

```python
from math import factorial
import numpy as np

def binomial(n, p, k):
    C = (factorial(n) / (factorial(n - k) * factorial(k)))
    return C * np.power(p, k) * np.power(1 - p, n - k)
```

```python
# probabilidade de sucesso (saque no caixa)
p = 0.4

# numero de clientes no caixa
n = 30

#frequencia observada (numero max de clientes sacando)
k = 12

# usando o for para fazer a somatória
prob = 0
for k in range(0, k+1):
    prob += binomial(n, p, k)

#resultado
print(f"P(0<=k<=12): {np.round(prob*100,1)} %")
```
