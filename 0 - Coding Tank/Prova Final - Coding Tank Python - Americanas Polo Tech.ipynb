Um parque de diversões tem 3 atrações principais: o carrossel, piscina de bolinhas e montanha-russa.

Para poder participar de uma atração a pessoa deve cumprir as seguintes condições:

Carrossel: altura mínima de 1,00m e idade mínima de 3 anos.

Piscina de bolinhas: idade entre 4 e 9 anos e máximo de 1,30m de altura.

Montanha-russa: altura mínima de 1,10m.
O fiscal de cada atração verificará o ano de nascimento da pessoa e altura para liberar o acesso para uma pessoa.

Faça uma função em python que receba o ano de nascimento e altura da pessoa e informe quais as atrações que a pessoa pode participar.

```python
altura = float(input('Altura da criança (Ex: 1.30): '))
nascimento = int(input('Ano de nascimento (Ex: 1992): '))

idade = 2022 - nascimento

def calculo (altura,idade):

  brinquedos = []
  brincar = False #tag pra validacao

  if altura >= 1 and idade >= 3:
      brincar = True
      brinquedos.append('Carrosel')
      
  
  if altura >= 1.3 and idade >= 4 and idade <= 9:
      brincar = True
      brinquedos.append('Piscina de Bolinhas') 


  if altura >= 1.1:
     brincar = True
     brinquedos.append('Montanha-Russa')
 
  
  if not brincar:
    brinquedos.append('Não vai brincar')

  return brinquedos 

calculo(altura,idade) #passando os parametros do input
```

Peça ao usuário para entrar com um número e faça uma função que retorne o fatorial dele como resposta. O fatorial de um número é o resultado da multiplicação de todos os números que o antecedem a partir de 1 até o número fornecido. Atenção, fatorial só vale para números positivos!

Exemplo: fatorial de 4 = 1 * 2 * 3 * 4 = 24

```python
numero_fatorial = int(input('Numero fatorial positivo: '))


def funcao_fatorial(numero_fatorial):

  contador_fatorial = 1

  if numero_fatorial > 0:
    while numero_fatorial > 0:
      contador_fatorial *=  numero_fatorial
      numero_fatorial -= 1

    return contador_fatorial

  else: return 'Só podemos fatorar números positivos'

funcao_fatorial(numero_fatorial) #passando os parametros do input
```

Dado um número fornecido pelo usuário, faça um programa que teste se é número primo e imprima na tela. Além disso, exiba uma lista de divisores do número testado. Um número primo é divisível somente por 1 e por ele mesmo. Seu programa deve ser funcional para qualquer número até o 100.

13 é divisível somente por 1 e ele mesmo (13), então é primo.

25 é divisível por 1, 5 e ele mesmo (25), então NÃO é primo.




```python
numero = int(input('Digite um numero para verificar se é primo: '))
divisores = [] 


if numero > 1:
    for i in range(2, numero+1): #comeca do 2 ja que 1 nao eh primo, o segundo parametro eh intervalo aberto, coloquei +1 pra pegar o numero em si
        if numero % i == 0:
              divisores.append(i) #adiciono na lista aqueles numeros do range que tem como resto 0
              
    if not divisores:  #criei isso daqui pensando na tag true/false e vi no stackoverflow que posso verificar uma lista vazia assim 
          print(f' {numero} é divisível somente por 1 e ele mesmo ({numero}), então é primo.') #nao sei como tirar o colchete do print
    else:
          print(f'{numero} é divisivel por {divisores} e ele mesmo ({numero}), então NÃO é primo.')

elif numero <= 0:
      print(f' {numero} não é um numero válido')
else:
      print(f' {numero} não é primo') #numero 1 nao eh primo
```

Faça um script que leia 10 números do usuário e informe se algum número foi inserido em sequência. Se mais de um número for repetido, informe ao menos um (caso no Exemplo 3).

Exemplo 1

input 1: 3

input 2: 5

(etc...) 6.. 7.. 12.. 2... 43.. 5.. 

input 9: 1

input 10: 8

output: Nenhum número foi repetido em sequência


Exemplo 2

input 1: 9

input 2: 4

(etc...) 9.. 5.. 9.. 6... 3.. 1.. 

input 9: 6

input 10: 6

output: O número 6 foi repetido em sequência

Exemplo 3

input 1: 2

input 2: 3

(etc...) 9.. 9.. 4.. 6... 7.. 7.. 

input 9: 4

input 10: 3

output: O número 7 foi repetido em sequência

```python
sequencia = []
numero_repetido = []
teste_sequencia = False #tag

for i in range(10): #preenchendo a lista de sequencia com os numeros
    numero_sequencia = int(input("Digite um numero: "))
    sequencia.append(numero_sequencia) 

for i in range(1, 10): #agora fazendo os calculos depois da lista de sequencia preenchida
    if(sequencia[i] - sequencia[i-1] == 0): #Valor da posicao atual - valor da posicao anterior == 0
        numero_repetido.append(sequencia[i]) #adiciona na lista de repetidos o valor desse numero da posicao atual, mas só ta puxando o primeiro repetido
        print(f'O(s) número(s) {numero_repetido} foi/foram digitado(s) em sequencia')
        teste_sequencia = True
        break

if teste_sequencia == False:
    print('Nenhum número foi repetido em sequencia')
```
